```r
library(dplyr)
library(Seurat)
library(patchwork)
library(harmony)

# remove ATAC assay
merged_afib[['ATAC']] <- NULL

# Harmonize based on RNA assay
merged_afib <- RunHarmony(
    object = merged_afib,
    group.by.vars = 'orig.ident',
    reduction = 'pca',
    assay.use = 'RNA',
    project.dim = FALSE,
    reduction.save = "harmony_rna"
)

#saveRDS(merged_afib, file = "/volume/AFproj/samples/merged_afib_postHarmony.rds")
```

```r
library(dplyr)
library(Seurat)
library(patchwork)
#merged_afib <- readRDS("/volume/AFproj/samples/merged_afib_postHarmony.rds")
merged_afib <- readRDS("/volume/AFproj/samples/merged_afib_postHarmony2.rds")

DefaultAssay(merged_afib) <- 'RNA'
merged_afib <- FindClusters(merged_afib)
merged_afib <- FindNeighbors(merged_afib, dims = 1:42)
merged_afib <- RunUMAP(merged_afib, reduction = 'harmony_rna', dims = 1:50, reduction.name = "umap.harmony.rna", reduction.key = "rnaUMAPHarmony_")
```

```r
merged_afib$state2 <- paste(merged_afib$condition, merged_afib$group)
merged_afib$state <- gsub("NA", "", merged_afib$state2)
merged_afib$state2 <- NULL
merged_afib$state <- gsub("CTRL", "ctrl_afib",merged_afib$state)
merged_afib$state <- gsub(" ctrl", "ctrl_naehrendorf",merged_afib$state)
merged_afib$state2 <- paste(merged_afib$state)
merged_afib$state2 <- gsub("ctrl_afib|ctrl_naehrendorf", "ctrl",merged_afib$state2)
merged_afib$state2 <- gsub(" mraf", "aFib",merged_afib$state2)


merged_afib@meta.data
unique(merged_afib$state2)
merged_afib$state2 <- trimws(merged_afib$state2)
unique(merged_afib$state2)
```

```r
# Subset the data based on the "celltype" column
CM <- merged_afib[, merged_afib$celltype == "aCM", ]
CM$percent.mt <- NA
CM@meta.data
aCM_rows <- afib_dataset$clusters2 == "aCM"
CM$percent.mt[aCM_rows] <- afib_dataset$percent.mt[aCM_rows]
```

```r
CM@meta.data
```

```r
# Violin Plot
CM[["percent.mt"]] <- PercentageFeatureSet(CM, pattern = "^MT-")

VlnPlot(CM, features = c("nFeature_RNA", "nCount_RNA", "percent.mt"), ncol = 3)

CM <- subset(CM, subset = nFeature_RNA > 200 & nFeature_RNA < 6500 & percent.mt < 4)

# Normalize
CM <- NormalizeData(CM)

# SCALE & PCA 
all.genes <- rownames(CM)
CM <- ScaleData(CM, features = all.genes)

CM <- FindVariableFeatures(CM, selection.method = "vst", nfeatures = 2000)
CM <- RunPCA(CM, features = VariableFeatures(object = CM))

# ELBOWPLOT
ElbowPlot(CM, ndims = 50)
```

```r
library(dplyr)
library(Seurat)
library(patchwork)
library(harmony)

# Harmonize based on RNA assay
CM <- RunHarmony(
    object = CM,
    group.by.vars = 'orig.ident',
    reduction = 'pca',
    assay.use = 'RNA',
    project.dim = FALSE,
    reduction.save = "harmony_rna"
)

DefaultAssay(CM) <- 'RNA'
CM <- RunUMAP(CM, reduction = 'harmony_rna', dims = 1:50, reduction.name = "umap.harmony.rna", reduction.key = "rnaUMAPHarmony_")

# UMAP
CM <- FindNeighbors(CM, dims = 1:42)
CM <- FindClusters(CM, resolution = 0.5)

pdf("/volume/AFproj/CM_UMAPS.pdf")
    print(DimPlot(CM,reduction = "umap.harmony.rna", group.by='orig.ident'))
    print(DimPlot(CM,reduction = "umap.harmony.rna", group.by='seurat_clusters'))
    print(DimPlot(CM,reduction = "umap.harmony.rna", group.by='state'))
    print(DimPlot(CM,reduction = "umap.harmony.rna", split.by='state'))
dev.off()
```

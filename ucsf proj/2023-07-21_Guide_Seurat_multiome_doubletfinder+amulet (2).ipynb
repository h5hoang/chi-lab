# Remake Seurat objects on scrublet + amulet corrected data

```r
# Load in the postQCfilt.rds objects and remove the doublets from scrublet and amulet
```

```r
for (i in seq_along(sampleNames)){
    sampleName <- sampleNames[i]
    seurat <- readRDS(paste0(out_dir,sampleName,'_postQCfilt.rds'))
    #sampleFile_rna <- paste0(inputFiles_dir[[i]],sampleName,'_soupX/')
    sampleFile_atac <- paste0(inputFiles_dir,sampleName,'_atac_fragments.tsv.gz')
    print(paste0('Removing doublets: ', sampleName))
    
    # Scrublet
    # Load in scrublet results after running the python script on soupx counts and add to metadata
    scrublet <- read.table(paste0(out_dir,sampleName,'_scrublet/scrublet_results.tsv'), sep='\t', header=TRUE)
    print('scrublet results:')
    print(table(scrublet$scrublet_DropletType))
    # seurat metadata cellnames already in the same order as scrublet result cellnames so we can just add the metadata in
    seurat@meta.data$scrublet_DropletType <- scrublet$scrublet_DropletType
    seurat@meta.data$scrublet_Scores <- scrublet$scrublet_Scores
    
    # AMULET
    # Load in Amulet results after running the bash script on celranger fragments and add to metadata
    amulet <- read.table(paste0(out_dir,sampleName,'_amulet/MultipletBarcodes_01.txt'), sep='\t', header=FALSE)
    seurat@meta.data$amulet_doubletType <- 'singlet'
    for (cell in amulet$V1) {
        if (cell %in% rownames(seurat@meta.data)) { # need to check if cell exists in our seurat object because amulet was run on all fragments from cellranger output
            seurat@meta.data[cell,'amulet_doubletType'] <- 'doublet'
        }
    }
    print('amulet results:')
    print(table(seurat@meta.data$amulet_doubletType))
    
    # create all_doublets metadata for both scrublet+amulet
    # get the cells that were deemed a doublet by scrublet or amulet
    cells <- rownames(seurat@meta.data %>% dplyr::filter(scrublet_DropletType == 'doublet' | amulet_doubletType == 'doublet'))
    seurat@meta.data$all_doublets <- 'singlet'
    for (cell in cells) {
        if (cell %in% rownames(seurat@meta.data)) { # need to check if cell exists in our seurat object (even though it is, good practice to check)
            seurat@meta.data[cell,'all_doublets'] <- 'doublet'
        }
    }
    print('total doublets to remove (scrublet + amulet):')
    print(table(seurat@meta.data$all_doublets))
    
    pdf(paste0(out_dir,'UMAPs_',sampleName,'_doublets.pdf'), width=7, height=7)
    print(DimPlot(seurat, group.by='scrublet_DropletType', reduction = "umap.rna"))
    print(DimPlot(seurat, group.by='amulet_doubletType', reduction = "umap.atac"))
    print(DimPlot(seurat, group.by='all_doublets', reduction = "umap.wnn"))
    dev.off()
       
    # Filter out the doublets
    Idents(seurat) <- 'all_doublets'
    seurat_filtered <- subset(seurat, idents='singlet')
        
    # Filter atac counts matrix to the filtered cells and remake the chromatin assay (possibly unnecessary, not sure)
    atac_counts <- seurat_filtered@assays$ATAC@counts
    #atac_counts <- atac_counts[,colnames(seurat_filtered)]
    
    grange.counts <- StringToGRanges(rownames(atac_counts), sep = c(":", "-"))
    grange.use <- seqnames(grange.counts) %in% standardChromosomes(grange.counts)
    atac_counts <- atac_counts[as.vector(grange.use), ]
    annotations <- GetGRangesFromEnsDb(ensdb = EnsDb.Hsapiens.v86)
    seqlevelsStyle(annotations) <- 'UCSC'
    genome(annotations) <- "hg38"

    chrom_assay <- CreateChromatinAssay(
       counts = atac_counts,
       sep = c(":", "-"),
       genome = 'hg38',
       fragments = sampleFile_atac,
       min.cells = 10,
       annotation = annotations
     )
    seurat_filtered[["ATAC"]] <- chrom_assay
    
    print(paste0('After doublet filtering: ', sampleName))
    print(seurat_filtered)

    filtered_list[[sampleName]] <- seurat_filtered
}
```

```r
# seurat pipeline on each object
for (i in seq_along(filtered_list)){
    sampleName <- names(filtered_list[i])
    seurat <- filtered_list[[i]]
    print(paste0('Reprocessing: ', sampleName))
    
    # RNA analysis
    DefaultAssay(seurat) <- 'RNA'
    seurat <- NormalizeData(seurat, verbose = FALSE)
    seurat <- FindVariableFeatures(seurat, verbose = FALSE)
    seurat <- ScaleData(seurat, verbose = FALSE)
    seurat <- RunPCA(seurat, verbose = FALSE)
    seurat <- RunUMAP(seurat, dims = 1:50, reduction.name = 'umap.rna', reduction.key = 'rnaUMAP_')
    
    # ATAC analysis
    # We exclude the first dimension as this is typically correlated with sequencing depth
    DefaultAssay(seurat) <- 'ATAC'
    seurat <- RunTFIDF(seurat)
    seurat <- FindTopFeatures(seurat, min.cutoff='q0')
    seurat <- RunSVD(seurat)
    seurat <- RunUMAP(seurat, reduction='lsi', dims=2:50, reduction.name='umap.atac', reduction.key='atacUMAP_')
    
    # WNN multimodal analysis, representing a weighted combination of RNA and ATAC-seq modalities. 
    seurat <- FindMultiModalNeighbors(seurat, reduction.list=list('pca', 'lsi'), dims.list=list(1:50, 2:50))
    seurat <- RunUMAP(seurat, nn.name='weighted.nn', reduction.name='umap.wnn', reduction.key='wnnUMAP_')
    seurat <- FindClusters(seurat, graph.name='wsnn', verbose=FALSE)
    
    pdf(paste0(out_dir,'UMAPs_',sampleName,'_postDoublet+RNAcontamFilt.pdf'), height=7, width=7)
    print(DimPlot(seurat, reduction = "umap.rna", label=TRUE, repel=TRUE))
    print(DimPlot(seurat, reduction = "umap.atac", label=TRUE, repel=TRUE))
    print(DimPlot(seurat, reduction = "umap.wnn", label=TRUE, repel=TRUE))
    dev.off()

    filtered_list[[sampleName]] <- seurat
    
    saveRDS(seurat, paste0(out_dir,sampleName,'_postDoublet+RNAcontamFilt.rds'))
}
```

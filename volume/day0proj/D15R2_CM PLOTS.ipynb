```r
library(dplyr)
library(Seurat)
library(patchwork)

# importing data 
D15R2_CM.data <- Read10X(data.dir = "/volume/DAY_0/D15R2_CM/D15R2_CM_DATA")
D15R2_CM <- CreateSeuratObject(counts = D15R2_CM.data, project = "D15R2_CM", min.cells = 3, min.features = 200)
D15R2_CM

# Violin Plot
D15R2_CM[["percent.mt"]] <- PercentageFeatureSet(D15R2_CM, pattern = "^MT-")

VlnPlot(D15R2_CM, features = c("nFeature_RNA", "nCount_RNA", "percent.mt"), ncol = 3)

# SUBSET DATA
D15R2_CM <- subset(D15R2_CM, subset = nFeature_RNA > 1250 & nFeature_RNA < 9000 & percent.mt < 50)

# SAVE
saveRDS(D15R2_CM, file = "/volume/DAY_0/D15R2_CM/D15R2_CM_final.rds")
```

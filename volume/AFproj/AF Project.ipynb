```r
library(dplyr)
library(Seurat)
library(patchwork)

# importing data 
GSM5919340.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919340_5_CTRL")
GSM5919341.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919341_6_CTRL")
GSM5919342.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919342_10_CTRL")
GSM5919343.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919343_11_CTRL")
GSM5919344.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919344_12_CTRL")
GSM5919345.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919345_1_MR_AF")
GSM5919346.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919346_2_MR_AF")
GSM5919347.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919347_3_MR_AF")
GSM5919348.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919348_4_MR_AF")
GSM5919349.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919349_7_MR_AF")
GSM5919350.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919350_13_MR_AF")
GSM5919351.data <- Read10X(data.dir = "/volume/AFproj/samples/GSM5919351_14_MR_AF")

GSM5919340 <- CreateSeuratObject(counts = GSM5919340.data, project = "Human Control 1", min.cells = 3, min.features = 200)
GSM5919341 <- CreateSeuratObject(counts = GSM5919341.data, project = "Human Control 2", min.cells = 3, min.features = 200)
GSM5919342 <- CreateSeuratObject(counts = GSM5919342.data, project = "Human Control 3", min.cells = 3, min.features = 200)
GSM5919343 <- CreateSeuratObject(counts = GSM5919343.data, project = "Human Control 4", min.cells = 3, min.features = 200)
GSM5919344 <- CreateSeuratObject(counts = GSM5919344.data, project = "Human Control 5", min.cells = 3, min.features = 200)
GSM5919345 <- CreateSeuratObject(counts = GSM5919345.data, project = "Human MR+AF 1", min.cells = 3, min.features = 200)
GSM5919346 <- CreateSeuratObject(counts = GSM5919346.data, project = "Human MR+AF 2", min.cells = 3, min.features = 200)
GSM5919347 <- CreateSeuratObject(counts = GSM5919347.data, project = "Human MR+AF 3", min.cells = 3, min.features = 200)
GSM5919348 <- CreateSeuratObject(counts = GSM5919348.data, project = "Human MR+AF 4", min.cells = 3, min.features = 200)
GSM5919349 <- CreateSeuratObject(counts = GSM5919349.data, project = "Human MR+AF 5", min.cells = 3, min.features = 200)
GSM5919350 <- CreateSeuratObject(counts = GSM5919350.data, project = "Human MR+AF 6", min.cells = 3, min.features = 200)
GSM5919351 <- CreateSeuratObject(counts = GSM5919351.data, project = "Human MR+AF 7", min.cells = 3, min.features = 200)

# MERGE
merged_samples_PRE_QC <- merge(GSM5919340, y = c(GSM5919341, GSM5919342, GSM5919343, GSM5919344, GSM5919345,GSM5919346,GSM5919347,GSM5919348,
                                         GSM5919349,GSM5919350,GSM5919351), add.cell.ids = c("CTRL1", "CTRL2", "CTRL3", "CTRL4", "CTRL5", 
                                                                                             "MR+AF1","MR+AF2","MR+AF3","MR+AF4","MR+AF5","MR+AF6","MR+AF7"), project = "merged_samples")
#saveRDS(merged_samples_PRE_QC, file = "/volume/AFproj/samples/merged_samples_PRE_QC.rds")
```

```r
merged_samples_PRE_QC <- readRDS("/volume/AFproj/samples/merged_samples_PRE_QC.rds")

# Violin Plot
merged_samples_PRE_QC[["percent.mt"]] <- PercentageFeatureSet(merged_samples_PRE_QC, pattern = "^MT-")

VlnPlot(merged_samples_PRE_QC, features = c("nFeature_RNA", "nCount_RNA", "percent.mt"), ncol = 3)

merged_samples_PRE_QC <- subset(merged_samples_PRE_QC, subset = nFeature_RNA > 200 & nFeature_RNA < 4500 & percent.mt < 25)

# Normalize
merged_samples_PRE_QC <- NormalizeData(merged_samples_PRE_QC)

# SCALE & PCA 
all.genes <- rownames(merged_samples_PRE_QC)
merged_samples_PRE_QC <- ScaleData(merged_samples_PRE_QC, features = all.genes)

merged_samples_PRE_QC <- FindVariableFeatures(merged_samples_PRE_QC, selection.method = "vst", nfeatures = 2000)
merged_samples_PRE_QC <- RunPCA(merged_samples_PRE_QC, features = VariableFeatures(object = merged_samples_PRE_QC))

# ELBOWPLOT
ElbowPlot(merged_samples_PRE_QC, ndims = 50)

#saveRDS(merged_samples_PRE_QC, file = "/volume/AFproj/samples/merged_samples_POST_QC.rds")
```

```r
# Creating conditions metadata column
names <- merged_samples_PRE_QC@meta.data$orig.ident
names <- gsub("1|2|3|4|5|6|7", "", names)
merged_samples_PRE_QC@meta.data$samplename <- names

condition <- merged_samples_PRE_QC@meta.data$samplename
condition <- gsub("Human Control", "control", condition)
condition <- gsub("Human MR\\+AF", "AFIB", condition)
merged_samples_PRE_QC@meta.data$condition <- condition

# METADATA Table
merged_samples_PRE_QC@meta.data

# UMAP
merged_samples_PRE_QC <- FindNeighbors(merged_samples_PRE_QC, dims = 1:42)
merged_samples_PRE_QC <- FindClusters(merged_samples_PRE_QC, resolution = 0.5)
merged_samples_PRE_QC <- RunUMAP(merged_samples_PRE_QC, dims = 1:42)

pdf("/volume/AFproj/Merged_UMAPS.pdf")
    print(DimPlot(merged_samples_PRE_QC, group.by='orig.ident'))
    print(DimPlot(merged_samples_PRE_QC, group.by='seurat_clusters'))
    print(DimPlot(merged_samples_PRE_QC, group.by='condition'))
dev.off()
```

# HARMONIZING DATA

```r
library(dplyr)
library(Seurat)
library(patchwork)
library(harmony)

merged_samples_POST_QC <- readRDS("/volume/AFproj/samples/merged_samples_POST_QC.rds")

# Harmonize based on RNA assay
merged_samples_POST_QC <- RunHarmony(
    object = merged_samples_POST_QC,
    group.by.vars = 'orig.ident',
    reduction = 'pca',
    assay.use = 'RNA',
    project.dim = FALSE,
    reduction.save = "harmony_rna"
)

saveRDS(merged_samples_POST_QC, file = "/volume/AFproj/samples/merged_samples_POST_QC+HARMONY.rds")
```

```r
DefaultAssay(merged_samples_POST_QC) <- 'RNA'
merged_samples_POST_QC <- FindClusters(merged_samples_POST_QC, resolution = 0.1)
merged_samples_POST_QC <- FindNeighbors(merged_samples_POST_QC, dims = 1:42)
merged_samples_POST_QC <- RunUMAP(merged_samples_POST_QC, reduction = 'harmony_rna', dims = 1:50, reduction.name = "umap.harmony.rna", reduction.key = "rnaUMAPHarmony_")
```

```r
pdf("/volume/AFproj/Merged_UMAPS_postHarmony.pdf")
    print(DimPlot(merged_samples_POST_QC, reduction = "umap.harmony.rna", group.by = "orig.ident", label = TRUE,repel = TRUE, shuffle=TRUE))
    print(DimPlot(merged_samples_POST_QC, reduction = "umap.harmony.rna", group.by = "seurat_clusters", label = TRUE, repel = TRUE,shuffle=TRUE))
    print(DimPlot(merged_samples_POST_QC, reduction = "umap.harmony.rna", group.by = "condition", label = TRUE,repel = TRUE, shuffle=TRUE))
    print(DimPlot(merged_samples_POST_QC, reduction = "umap.harmony.rna", group.by = "condition", split.by = "condition", shuffle=TRUE))
dev.off()
```
